# **Prueba de hipotesis**

**1. Definición de hipótesis**

Dada las simetrias medias que se obtienen de las muestras, se evidencia que las muestras no son pareadas porque la cantidad de datos es distnta.

**H0= Ub < Um**

**H1= Ub > Um** 

Hipotesis nula la media de simetria para celulas benignas es menor que la media de simetria para celulas malignas.

Hipotesis alternativa la media de simetria para celulas benignas es mayor que la media de simetria para celulas malignas.


**2. Comprobación de supuestos**


**H₀: Los datos se distribuyen normalmente.**

**H₁: los datos no se distribuyen normalmente.**

Suponga que α=0.05. Si el valor p es >0,05, se puede decir que los datos se distribuyen normalmente.

https://towardsdatascience.com/hypothesis-testing-with-python-step-by-step-hands-on-tutorial-with-practical-examples-e805975ea96e

```python
def check_normality(data): 
    test_stat_normality, p_value_normality=sc.stats.shapiro(data)
    print("p value:%.4f" % p_value_normality)
    if p_value_normality <0.05:
        print("Rechazar la hipótesis nula >> Los datos no se distribuyen normalmente")
    else:
        print("No se puede rechazar la hipótesis nula >> Los datos se distribuyen normalmente")
```

```python
check_normality(symetry_b) # Me indica si los datos son normales.
check_normality(symetry_m)
```

**Evidenciamos que los datos no se distribuyen normalmente, esto es que la simetria media benigna no es normal y la simetria media maligna no es normal.**

```python
def check_variance_homogeneity(group1, group2): # Prueba de varianza
    test_stat_var, p_value_var= sc.stats.levene(group1,group2)
    print("p value:%.4f" % p_value_var)
    if p_value_var <0.05:
        print("Rechazar hipótesis nula >> Las varianzas de las muestras son diferentes.")
    else:
        print("No se puede rechazar la hipótesis nula >> Las varianzas de las muestras son las mismas.")
```

```python
check_variance_homogeneity(symetry_b, symetry_m)
```

```python
ttest,p_value = sc.stats.ttest_ind(symetry_b, symetry_m) # Prueba de independencia
print("p value:%.8f" % p_value)
print("Ya que la hipótesis es unilateral >> use p_value/2 >> p_value_one_side:%.4f" %(p_value/2))
if p_value/2 <0.05:
    print("Rechazar hipótesis nula")
else:
    print("No se puede rechazar la hipótesis nula")
```

**La primera prueba t Tes no cumplio con los requisitos, por lo cual se debe realizar otra Mann Whitney U.**

# **Mann Whitney U**

```python
check_normality(symetry_b)
check_normality(symetry_m)
check_variance_homogeneity(symetry_b, symetry_m)
```

```python
ttest,pvalue = sc.stats.mannwhitneyu(symetry_b, symetry_m, alternative="two-sided")
print("p-value:%.4f" % pvalue)
if pvalue <0.05:
    print("Rechazar hipótesis nula")
else:
    print("No se puede recibir la hipótesis nula")
```

**Debido a que los datos no cumplen ciertos requisitos para ser evaluados a través de una prueba paramétrica, se realizo Mann whitney U y la hipotesis nula debe ser rechazada.**
